```python
# Personal Finance Tracker with Exception Handling

try:
    income = float(input("Enter your monthly income: "))

    n = int(input("How many expenses? "))
    expenses = []

    for i in range(n):
        amt = float(input(f"Enter expense {i+1}: "))
        expenses.append(amt)

    total_expense = sum(expenses)
    savings = income - total_expense
    percent_spent = (total_expense / income) * 100


    if percent_spent < 40:
        category = "LOW"
        message = "Good control!"
    elif percent_spent < 70:
        category = "MODERATE"
        message = "Be careful!"
    else:
        category = "HIGH"
        message = "Financial Risk!"

    print("\n--- FINANCE SUMMARY ---")
    print("Total Expenses:", total_expense)
    print("Savings:", savings)
    print("Percent Spent:", percent_spent)
    print("Category:", category)
    print("Message:", message)

except Exception as e:
    print("Error occurred:", e)


```

```python
#ques 1
# Demonstrating Inheritance + Encapsulation + Polymorphism

class Animal:
    def __init__(self, name, age):
        self.name = name
        self.__age = age

    def get_age(self):
        return self.__age

    def make_sound(self):
        pass


class Dog(Animal):
    def make_sound(self):
        return "Bark!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"


animals = [Dog("Tommy", 5), Cat("Whiskers", 3)]

for a in animals:
    print(f"{a.name} says {a.make_sound()} and is {a.get_age()} years old")
```

```python
#ques 2
# Polymorphism Example: Ride Fare Calculation

class Vehicle:
    def calculate_fare(self, distance):
        pass

class Auto(Vehicle):
    def calculate_fare(self, distance):
        return 30 + distance * 10

class Bike(Vehicle):
    def calculate_fare(self, distance):
        return 20 + distance * 7

class Mini(Vehicle):
    def calculate_fare(self, distance):
        return 50 + distance * 12


rides = [Auto(), Bike(), Mini()]

distance = 10

for ride in rides:
    print(f"{ride.__class__.__name__} Fare: ₹{ride.calculate_fare(distance)}")
```

```python
# ques 3
# Base Class
class Staff:
    def __init__(self, name, staff_id):
        self.name = name
        self.staff_id = staff_id

    def show_details(self):
        print(f"Name: {self.name}, ID: {self.staff_id}")


class Doctor(Staff):
    def treat_patient(self):
        print(f"Dr. {self.name} is treating a patient.")


class Nurse(Staff):
    def assist(self):
        print(f"Nurse {self.name} is assisting in the ward.")

class Surgeon(Staff):
    def perform_surgery(self):
        print(f"Surgeon {self.name} is performing surgery.")


class LabTechnician(Staff):
    def test_samples(self):
        print(f"Lab Tech {self.name} is testing samples.")

staff_list = [
    Doctor("Rahul", 101),
    Nurse("Priya", 102),
    Surgeon("Aman", 103),
    LabTechnician("Sana", 104)
]


for member in staff_list:
    member.show_details()

    if isinstance(member, Doctor):
        member.treat_patient()
    elif isinstance(member, Nurse):
        member.assist()
    elif isinstance(member, Surgeon):
        member.perform_surgery()
    elif isinstance(member, LabTechnician):
        member.test_samples()
```

```python
#ques 4
class BankAccount:
    def __init__(self, name, balance, pin, credit_limit):
        self.name = name
        self.__balance = balance
        self.__pin = pin
        self.__credit_limit = credit_limit


    def get_balance(self, pin):
        if pin == self.__pin:
            return self.__balance
        else:
            return "Access Denied: Invalid PIN"


    def deposit(self, amount):
        self.__balance += amount
        return f"Deposited {amount}. New Balance: {self.__balance}"


    def withdraw(self, amount, pin):
        if pin != self.__pin:
            return "Invalid PIN"
        if amount > self.__balance:
            return "Insufficient Funds"
        self.__balance -= amount
        return f"Withdrawal successful. New Balance: {self.__balance}"


acc = BankAccount("Ruchi", 5000, 1234, 20000)

print(acc.get_balance(1234))
print(acc.deposit(2000))
print(acc.withdraw(3000, 1234))
print(acc.get_balance(1111))
```

```python
#ques 5
class StudentRecord:
    def __init__(self, name, reg_no, internal_marks, medical_leaves, discipline_status):
        self.name = name
        self.reg_no = reg_no
        self.__internal_marks = internal_marks
        self.__medical_leaves = medical_leaves
        self.__discipline_status = discipline_status

    def get_internal_marks(self, role):
        if role == "faculty":
            return self.__internal_marks
        else:
            return "Access Denied"


    def set_internal_marks(self, marks, role):
        if role != "faculty":
            return "Unauthorized Update"
        if 0 <= marks <= 30:
            self.__internal_marks = marks
            return "Marks Updated Successfully"
        return "Invalid Marks"

    def get_medical_leaves(self, role):
        return self.__medical_leaves if role == "faculty" else "Access Denied"

    def get_discipline_status(self, role):
        return self.__discipline_status if role == "admin" else "Access Denied"



student = StudentRecord("Ruchi", 101, 25, 3, "No issues")

print(student.get_internal_marks("faculty"))
print(student.set_internal_marks(28, "faculty"))
print(student.get_internal_marks("student"))
print(student.get_discipline_status("admin"))
```

```python
#ques 6
from abc import ABC, abstractmethod


class PaymentMethod(ABC):

    @abstractmethod
    def pay(self, amount):
        pass


class UPI(PaymentMethod):
    def pay(self, amount):
        return f"Paid ₹{amount} using UPI."



class CreditCard(PaymentMethod):
    def pay(self, amount):
        return f"Paid ₹{amount} using Credit Card."



class Wallet(PaymentMethod):
    def pay(self, amount):
        return f"Paid ₹{amount} using Wallet balance."



payments = [UPI(), CreditCard(), Wallet()]

for p in payments:
    print(p.pay(500))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    "age": [20, 25, 30, np.nan, 40, 45],
    "bmi": [18, 20, 22, 23, 30, 35],
    "steps": [5000, 6000, 7000, np.nan, 4000, 3000]
})


print(df.isnull().sum())

df.fillna(df.mean(), inplace=True)

Q1 = df["bmi"].quantile(0.25)
Q3 = df["bmi"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["bmi"] < lower) | (df["bmi"] > upper)]
print("BMI Outliers:\n", outliers)

print(df.describe())

plt.hist(df["age"])
plt.title("Age Distribution")
plt.show()

sns.boxplot(df["bmi"])
plt.title("BMI Boxplot")
plt.show()

sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "amount": [200, 250, 300, 280, 270, 1000, 1200]
})

mean = df["amount"].mean()
std = df["amount"].std()

df["z"] = (df["amount"] - mean) / std
z_outliers = df[df["z"].abs() > 3]
print(z_outliers)

Q1 = df["amount"].quantile(0.25)
Q3 = df["amount"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

iqr_outliers = df[(df["amount"] < lower) | (df["amount"] > upper)]
print(iqr_outliers)

median = df["amount"].median()
df.loc[df["amount"] > upper, "amount"] = median
print(df)

```
